```python
def isValid(self, s: str) -> bool:
    # For c in s
    # If left bracket, append to stack
    # If right bracket, pop stack and compare
    # If violation, return False
    # Return True
    l = []
    dleft = {'(': ')',
             '{': '}',
             '[': ']',
            }
    for c in s:
        if c in dleft:
            l.append(c)
        else:
            try:
                if c != dleft[l.pop()]:
                    return False
            except:
                return False

    return len(l)==0
```

```python
# 10-7-22. Slightly faster
def isValid(self, s: str) -> bool:
    # 1) Make a list and append left brackets into it
    # 2) If a right bracket is hit, pop the last element of the stack
    # and check that the dictionary value of that token is correct
    # 3) If we reach end of s, and stack is not empty, then return false
    stack = []
    d = {')': '(', '}': '{', ']':'['}
    for c in s:
        if c not in d:
            stack.append(c) # if not a right bracket, then put on stack
        else:
            try:
                if d[c] == stack.pop():
                    pass
                else:
                    return False
            except:
                return False
    if len(stack) > 0:
        return False
    return True
```

```python
def isValid(s):
    l = []
    right_brackets = (')', ']', '}')
    d_check = {'(': ')', '[': ']', '{': '}'}
    if len(s) == 1:
        return False   

    for c in s:
        if c not in right_brackets:
            l.append(c)
        else:
            try:
                if c != d_check[l.pop()]:
                    return False
            except:
                return False

    if len(l) != 0:
        return False        
    return True
```
